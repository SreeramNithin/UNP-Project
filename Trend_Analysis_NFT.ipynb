```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("C:\\Users\\Nithin\\OneDrive\\Desktop\\abc.xlsx")
data
```

```python
from sklearn.linear_model import LinearRegression
```

```python
sns.pairplot(data.sample(20),x_vars=['Open','High','Low'],y_vars='Close',size=7 , aspect=0.7 , kind = 'reg')
```

```python
sns.pairplot(data.sample(20),x_vars=['Open1','High1','Low1'],y_vars='Close1',size=7 , aspect=0.7 , kind = 'reg')
```

```python
from sklearn.model_selection import train_test_split
import sklearn
import sklearn.metrics as met

x = data.drop(['Open','High','Low','Close','Close1'],axis=1)
y = data.Close1
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.30,random_state=42)
lm1 = LinearRegression()
lm1.fit(x_train,y_train)
y_pred = lm1.predict(x_test)
print(y_pred)
print(np.sqrt(met.mean_squared_error(y_test,y_pred)))

```
